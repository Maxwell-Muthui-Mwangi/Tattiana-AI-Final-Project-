```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 8))
ax.axis('off')

# Define the actor and its position
actor_position = (0.1, 0.8)
actor_radius = 0.05
ax.text(actor_position[0] - 0.05, actor_position[1], 'User', fontsize=12, ha='right', va='center')
actor_circle = plt.Circle(actor_position, actor_radius, color='black', fill=False)
ax.add_patch(actor_circle)

# Define use cases and their positions
use_cases = {
    "Content Moderation": (0.5, 0.8),
    "Fraud Detection": (0.5, 0.6),
    "Chatbot": (0.5, 0.4),
    "Translation": (0.5, 0.2)
}

# Draw use cases
for use_case, position in use_cases.items():
    ax.text(position[0], position[1], use_case, fontsize=10, ha='center', va='center', bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightblue'))
    ellipse = patches.Ellipse(position, width=0.3, height=0.1, edgecolor='black', facecolor='lightblue')
    ax.add_patch(ellipse)

# Draw lines connecting actor to use cases
for position in use_cases.values():
    ax.plot([actor_position[0] + actor_radius, position[0] - 0.15], [actor_position[1], position[1]], color='black')

# Show plot
plt.title("Use Case Diagram for Tattiana AI", fontsize=14)
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches

def draw_diagram(diagram_title, positions, sizes, arrows, additional_text=None):
    fig, ax = plt.subplots(figsize=(14, 8))

    # Add the title
    plt.title(diagram_title, fontsize=16)

    # Add the boxes to the plot
    for label, (x, y) in positions.items():
        width, height = sizes[label]
        ax.add_patch(patches.FancyBboxPatch((x, y), width, height, boxstyle="round,pad=0.02", edgecolor='black', facecolor='lightblue'))
        ax.text(x + width/2, y + height/2, label, ha='center', va='center', fontsize=10)

    # Add arrows between the boxes
    for start, end in arrows:
        start_x, start_y = positions[start]
        end_x, end_y = positions[end]
        start_width, start_height = sizes[start]
        end_width, end_height = sizes[end]

        arrow = patches.FancyArrowPatch((start_x + start_width, start_y + start_height/2),
                                        (end_x, end_y + end_height/2),
                                        connectionstyle="arc3,rad=0.2",
                                        arrowstyle='-|>', mutation_scale=20, color="black")
        ax.add_patch(arrow)

    # Add additional text if provided
    if additional_text:
        for label, (x, y) in additional_text.items():
            ax.text(x, y, label, ha='center', va='center', fontsize=10, bbox=dict(boxstyle="round,pad=0.3", edgecolor='black', facecolor='lightgray'))

    # Set the limits and hide the axes
    ax.set_xlim(0, 1)
    ax.set_ylim(0, 1)
    ax.axis('off')

    # Show the plot
    plt.show()

# Data for Fraud Detection Process diagram
positions_fd = {
    "User": (0.05, 0.6),
    "Input Transaction Data": (0.25, 0.6),
    "Fraud Detection Process": (0.5, 0.6),
    "Output": (0.9, 0.6),
    "Fraud Alerts": (0.55, 0.4),
    "Detection Report": (0.55, 0.2),
    "Output (Fraud Alerts)": (0.75, 0.4),
    "Output (Detection Report)": (0.75, 0.2)
}

sizes_fd = {
    "User": (0.1, 0.1),
    "Input Transaction Data": (0.2, 0.1),
    "Fraud Detection Process": (0.25, 0.1),
    "Output": (0.1, 0.1),
    "Fraud Alerts": (0.2, 0.05),
    "Detection Report": (0.2, 0.05),
    "Output (Fraud Alerts)": (0.1, 0.05),
    "Output (Detection Report)": (0.1, 0.05)
}

arrows_fd = [
    ("User", "Input Transaction Data"),
    ("Input Transaction Data", "Fraud Detection Process"),
    ("Fraud Detection Process", "Output"),
    ("Fraud Detection Process", "Fraud Alerts"),
    ("Fraud Detection Process", "Detection Report"),
    ("Fraud Alerts", "Output (Fraud Alerts)"),
    ("Detection Report", "Output (Detection Report)")
]

additional_text_fd = {
    "Fraud Alerts": (0.75, 0.425),
    "Detection Report": (0.75, 0.225)
}

# Draw the diagram for Fraud Detection Process
draw_diagram("Fraud Detection Process Data Flow", positions_fd, sizes_fd, arrows_fd, additional_text_fd)
```
